```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.info()
```

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
data.head()
```

```python
data['Geography'].unique()
```

```python
data['Geography'].value_counts()
data['Geography'] = data['Geography'].fillna('France')
data['Geography'].unique()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['HasCrCard'].value_counts()
data['HasCrCard'] = data['HasCrCard'].fillna(1.0)
data['HasCrCard'].value_counts()
```

```python
data['IsActiveMember'].value_counts()
data['IsActiveMember'] = data['IsActiveMember'].fillna(1.0)
data['IsActiveMember'].value_counts()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder.toarray()
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data['Gender']
```

```python
data.head()
```

```python
geo_encoder_df = pd.DataFrame(geo_encoder.toarray(), columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
data = pd.concat([data.drop('Geography', axis=1), geo_encoder_df], axis=1)
data.head()
```

```python
with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder_geo.pkl', 'wb') as file:
    pickle.dump(onehot_encoder_geo, file)
```

```python
x = data.drop('Exited', axis=1)
y = data['Exited']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
x_train
```

ANN IMPLEMENTATION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense #Dense means it is fully connected neural networks
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard #EarlyStopping is used to stop the training and monitoring process when the model's performance on a validation set starts to degrade, preventing overfitting.
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
early_stopping_callback
```

```python
import datetime
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
tensorflow_callback
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, validation_data=[x_test, y_test], epochs=100, callbacks=[early_stopping_callback, tensorflow_callback])
```

```python
model.save('model.h5')
```
